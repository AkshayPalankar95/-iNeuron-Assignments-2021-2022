#### 3. Compare and contrast map, filter, and reduce.

**Solution:** The differences between map, filter and reduce are: 

*> `map()`: The map() function is a type of higher-order. This function takes another function as a parameter
along with a sequence of iterables and returns an output after applying the function to each iterable present in the
sequence. 

*> `filter()`: The filter() function is used to create an output list consisting of values for which the function returns true.

*>. `reduce()`: The reduce() function, as the name describes, applies a given function to the iterables and returns a single value

```python
from functools import reduce
# map function
print('Map ->',list(map(lambda x:x*x, [1,2,3,4])))
# fitler function
print('Filter ->',list(filter(lambda x:x%2 !=0, [1,2,3,4])))
# reduce function
print('Reduce ->',reduce(lambda x,y:x+y, [1,2,3,4,5,6]))
```

#### 6. What are some general design guidelines for coding functions?
**Solution:** Some of the general design guidelines for coding functions are:
1. Always use a docstring to explain the functionality of the function
2. avoid using or limited use of global variables
3. Proper Identation to increase the code readability
4. try to follow a naming convention for function names (pascalCase or camelCase) and stick with the same convention throughout the application.
5. try to use a name for the function which conveys the purpose of the function
6. Local variables should be named using camelCase format `(ex: localVariable)` whereas Global variables names should be using PascalCase `(ex:GlobalVariable)`. 
7. Constant should be represented in allcaps `(ex:CONSTANT)`.
